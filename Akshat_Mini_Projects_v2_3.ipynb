```julia
len = length(Air_df[!,:RecentDelays])
function chunkIt(seq, num)
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq)
        out.append(seq[int(last):int(last + avg)])
        last += avg
      end
    return out
    end
```
